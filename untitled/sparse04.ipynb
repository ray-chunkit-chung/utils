```python
# from utils.pandas import read_text
import pandas as pd
import numpy as np
from io import StringIO

data = StringIO("""
    A   B   C     D  E
0  foo one small  1  1
1  foo one large  2  1
2  foo one large  2  1
5  bar one large  4  1
6  bar one small  5  1
7  bar two small  6  1
8  bar two large  7  1
9  foo th1 small  3  1
10  foo th2 small  3  1
11  foo th3 small  3  1
12  foo th4 small  3  1
""")

df = pd.read_csv(data,sep=r'\s+')
```

```python
# columns = None
columns = ['C','D']
index=['A', 'B']
# index=None
aggfunc=np.mean
# values=['E']
```

```python
from scipy.sparse import csr_matrix

def df_to_sm(data, vars_i, vars_j, aggfunc='mean'):
    grpr_i = data.groupby(vars_i).grouper
    idx_i = grpr_i.group_info[0]
    grpr_j = data.groupby(vars_j).grouper
    idx_j = grpr_j.group_info[0]
    
    print(grpr_i.group_info)
    print(grpr_i.groups)
    print()
    
    print(grpr_j.group_info)
    print(grpr_j.groups)

    data_sm = csr_matrix((data['E'].values, (idx_i, idx_j)),
                         shape=(grpr_i.ngroups, grpr_j.ngroups))

    return data_sm, grpr_i, grpr_j
```

```python
result, _, _ = df_to_sm(df,vars_i=index,vars_j=columns,aggfunc=aggfunc)

result.todense()
```
